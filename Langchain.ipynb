#### **Important Security Note:**

Storing API keys directly in files like `api_keys.yml` can be risky, especially if you share your notebook or push it to a public repository.

**Google Colab provides a secure way to store API keys using "Secrets".**

**Recommendation:**
1.  **Add your API key to Colab Secrets:**
    *   Click the "🔑" icon in the left sidebar.
    *   Click "Add a new secret".
    *   Set the **Name** to `OPENAI_API_KEY` (or whatever your API key is called).
    *   Paste your API key into the **Value** field.
    *   Ensure "Notebook access" is enabled for this notebook.
2.  **Access the secret in your code:** You can then access it using `from google.colab import userdata; OPENAI_API_KEY = userdata.get('OPENAI_API_KEY')`.
